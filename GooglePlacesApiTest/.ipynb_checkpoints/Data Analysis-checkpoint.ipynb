```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from textblob import TextBlob
```

```python

with open('place_details.json') as f:
    data = json.load(f)
    
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df['location'] = df['geometry'].apply(lambda x: x.get('location'))
df['lat'] = df['location'].apply(lambda x: x.get('lat'))
df['lon'] = df['location'].apply(lambda x: x.get('lng'))
```

```python
nyc = plt.imread('map.png')
BBox = (df.lon.min(),   df.lon.max(),      
         df.lat.min(), df.lat.max())
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df.lon-0.013, df.lat-0.05, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Bars and Restaurants in NYC')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(nyc, zorder=0, extent = BBox,)
plt.show()
```

```python
polar = []
for index, place in df.iterrows():
    polarity = 0.0 
    if type(place['reviews']) != float:
        for review in place['reviews']:
            sentiment = TextBlob(review['text']).sentiment.polarity
            polarity += sentiment
        polarity = polarity/len(place)
    polar.append(polarity)
```

```python
df['polarity'] = polar
```

```python
sns.set()
dfs = df[df['polarity'] != 0]
dfs = df[df['rating'] != 5]
ax = sns.scatterplot(x = 'rating', y = 'polarity', data = dfs)
```
